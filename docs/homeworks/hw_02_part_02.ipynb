````
# librerias
import glob

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
````

````
# Se empieza leyendo la data de entrenamiento
path = '/content/drive/MyDrive/Aplica/hw2/'
train_df = pd.read_csv(path + "train.csv")
train_df.head(5)
````

````
# Se establecer√° 'PassengerId' como el indice
train_df.set_index('PassengerId',inplace=True)
test_df.set_index('PassengerId',inplace=True)
````

# An√°lisis de datos


````
# Se revisar√°n valores NaN en los dataframes
print("Para conjunto de entrenamiento")
print(train_df.info())
print("")
print("Para conjunto de pruebas")
print(test_df.info())
````

Identificamos la presencia de valores NaN en las columnas 'Age' y 'Cabin' en los datos de entrenamiento. Similarmente, en el conjunto de pruebas, las columnas 'Age' y 'Cabin' tambi√©n contienen NaN, mientras que en 'Fare' se registra un √∫nico valor nulo.

Los NaN en 'Cabin' podr√≠an relacionarse con su aparici√≥n √∫nicamente en pasajeros con boletos de primera clase y escasamente en otras categor√≠as.

La ausencia de datos en 'Age' requerir√° una revisi√≥n detallada para entender por qu√© algunos pasajeros carecen de este registro.

Por √∫ltimo, el √∫nico valor nulo en 'Fare' ser√° analizado espec√≠ficamente para comprender su naturaleza.

````
train_df = train_df.drop(['Name','Cabin','Ticket'],axis=1)
test_df = test_df.drop(['Name','Cabin','Ticket'], axis=1)
````

````
# Se revisar√°n  valores NaN en los dataframes (nuevamente)
print("Para conjunto de entrenamiento")
print(train_df.info())
print("")
print("Para conjunto de pruebas")
print(test_df.info())
````

Dado que 'Name', 'Cabin' y 'Ticket' son columnas que tienen utilidad solo para identificar al pasajero o no aportan informacion util, estas se eliminan.

````
# Se revisa distribucion de valores categoricos en la columna 'Embarked' en cada conjunto
train_df['Embarked'].hist()
````

````
test_df['Embarked'].hist()
````



*   En el caso de los valores NaN de 'Age', 'Fare' estos ser√°n reemplazados por la media.
*   Para el caso de los valores NaN de 'Embarked', estos se reemplazar√°n por la clase de menor presencia.






````
train_df["Age"].fillna(train_df["Age"].mean(),inplace=True)
train_df["Embarked"].fillna("Q",inplace=True)

test_df["Age"].fillna(test_df["Age"].mean(),inplace=True)
test_df["Embarked"].fillna("Q",inplace=True)
test_df["Fare"].fillna(test_df["Fare"].mean(),inplace=True)
````

````
# Se volver√° a revisar presencia de valores NaN en los conjuntos
print("Para conjunto de entrenamiento")
print(train_df.info())
print("")
print("Para conjunto de pruebas")
print(test_df.info())
````

Una vez visto lo anterior se proceder√° con el preprocesamiento de los datos, los cuales son:

*  Sex y Embarked: Estos corresponden a variables categ√≥ricas, las cuales se les aplicar√° OneHotEncoder.

*  Age, SibSp, Parch y Fare: Estas variables son continuas, a las que se les aplicar√° StandardScaler.
*   Pclass: Se considera como variable categ√≥rica 'Ordinal', es decir, sus clases pueden ordenarse (en este caso son 1,2 y 3).






````
cat= ['Sex','Embarked']
cont = ['Age','SibSp','Parch','Fare']
train_df_cat = train_df[cat]
train_df_cont = train_df[cont]
train_df_pclass = train_df['Pclass']

test_df_cat = test_df[cat]
test_df_cont = test_df[cont]
test_df_pclass = test_df['Pclass']
````

````
# Se aplica encoder a datos categoricos
encoder1 = OneHotEncoder(sparse=False)
index_train_cat = train_df_cat.index
train_df_cat = encoder1.fit_transform(train_df_cat[cat])
train_df_cat = pd.DataFrame(train_df_cat,columns=encoder1.get_feature_names_out(cat))
train_df_cat.index = index_train_cat

index_test_cat = test_df_cat.index
encoder2 = OneHotEncoder(sparse=False)
test_df_cat = encoder2.fit_transform(test_df_cat[cat])
test_df_cat = pd.DataFrame(test_df_cat,columns=encoder2.get_feature_names_out(cat))
test_df_cat.index = index_test_cat
````

````
train_df_cat.head()
````

````
test_df_cat.head()
````

````
# Se escalan valores continuos
scaler1 = StandardScaler()
scaler1.fit(train_df_cont)
index_train_cont = train_df_cont.index
train_df_cont = pd.DataFrame(scaler1.transform(train_df_cont),columns=cont)
train_df_cont.index = index_train_cont

scaler2 = StandardScaler()
scaler2.fit(test_df_cont)
index_test_cont = test_df_cont.index
test_df_cont = pd.DataFrame(scaler2.transform(test_df_cont),columns=cont)
test_df_cont.index = index_test_cont
````

````
# Se concatenan los datos ya procesados
train_df_X = pd.concat([train_df_pclass,train_df_cont,train_df_cat],axis=1)
test_df = pd.concat([test_df_pclass,test_df_cont,test_df_cat],axis=1)
````

# Aplicaci√≥n del modelo


````
# Separamos los conjuntos
X = train_df_X
Y = train_df['Survived']

# Separamos los datos para el entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Usaremos un Random Forest
model = RandomForestClassifier(random_state=42)
model.fit(X_train,y_train)
````

````
# Evaluaci√≥n del modelo
y_pred = model.predict(X_test)
print("La precisi√≥n es: ",accuracy_score(y_test,y_pred))
````
