```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def towers_of_hanoi(n, source, target, auxiliary, moves=[]):
    if n == 1:
        moves.append((source, target))
        return moves
    towers_of_hanoi(n-1, source, auxiliary, target, moves)
    moves.append((source, target))
    towers_of_hanoi(n-1, auxiliary, target, source, moves)
    return moves

def visualize_towers(moves):
    pegs = {peg: list(range(n_disks, 0, -1)) for peg in ['A', 'B', 'C']}
    for move in moves:
        source, target = move
        disks = pegs[source]
        if disks:
            disk = disks.pop()
            pegs[target].append(disk)
            plt.clf()
            plt.bar(['A', 'B', 'C'], [len(pegs['A']), len(pegs['B']), len(pegs['C'])])
            plt.title('Towers of Hanoi')
            plt.xlabel('Peg')
            plt.ylabel('Number of Disks')
            plt.pause(0.5)
    plt.show()

n_disks = 3
source_peg = 'A'
target_peg = 'C'
auxiliary_peg = 'B'

moves = towers_of_hanoi(n_disks, source_peg, target_peg, auxiliary_peg)
visualize_towers(moves)
```
