```python
# read data
## products dataset
path = '/content/drive/MyDrive/Aplica/'
products_df = pd.read_csv(path + "products_info.csv")
products_df.columns = [x.lower().replace(' ','_') for x in products_df.columns]
products_df.head(5)
```

```python
## engagement dataset
path = '/content/drive/MyDrive/Aplica/engagement_data/'
all_files = glob.glob(path + "/*.csv")
li = []
all_files

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0,engine=None) #usamos engine=None para solucionar un error
    district_id = filename.split("/")[-1].split(".")[0]
    df["district_id"] = district_id
    li.append(df)

engagement_df = pd.concat(li)
engagement_df = engagement_df.reset_index(drop=True)
engagement_df
```

__ANALISIS DE DATOS: DISTRITOS__

Para comenzar, se analizarán los datos de los distritos. Para poder realizar esto, se hará un gráfico de "torta" para saber qué zonas fueron las más estudiadas.

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Distribución de estados', size = 20)
districts_df["state"].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
```

Notar que la mayor cantidad de personas estudiadas viven en los estados de Connecticut, Utah, Massachusetts e Illinois.

Sabiendo la distribución del estudio, será interesante estudiar el tipo de zona donde los encuestados están viviendo. Esto se realizará con el fin de tener un mejor análisis de datos.

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Distribución de Localidades', size = 20)
districts_df["locale"].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
plt.show()
```

Se puede observar que, a pesar de que un gran porcentaje de personas vive en lugares urbanos/suburbamos, hay un 18.8% de personas que viven en zonas rurales, esto podría traducirse como una dificultad al acceso a internet.

Analizando los distritos, se puede extraer más información. Podemos calcular qué personas se representan como negras/hispanohablantes, qué personas tienen becas alimenticias, esto pues, se les considera minorías.

```python
#Personas consideradas negras/hispanas
fig, ax  = plt.subplots(figsize=(16, 8))
sns.countplot(data= districts_df, x = "pct_black/hispanic");
plt.title("Black/hispanic")
plt.show();
```

```python
fig, ax  = plt.subplots(figsize=(16, 8))
sns.countplot(data= districts_df, x = "pct_free/reduced");
plt.title("Free/Reduced")
plt.show();
```

De esto se deduce que el 20% de las personas del estudio son negros/hispanos, y aproximadamente un 30%-40% de los estudiados se benefician de becas alimenticias

__ANALISIS DE DATOS: DISTRIBUIDORES__

Ya analizados los datos de los distritos, es momento de analizar los datos de la compañía. Primero, veamos cuáes son las compañías más potentes del mercado de aprendizaje online.

```python
plt.figure(figsize = (20,10))
sns.countplot(data=products_df, y="provider/company_name", order=products_df['provider/company_name'].value_counts().index[:10])
plt.title('Top 10 Provider Platform with the Most Product in 2020', size=12)
plt.show()
```

Podemos ver una notoria ventaja de Google LLC frente a los demás proveedores.

Ahora, veamos qué tipo de servicios son los más usados en la enseñanza.

```python
plt.figure(figsize = (16,8))
sns.countplot(x=products_df["primary_essential_function"]);
plt.xticks(rotation=90);
```

Se pueden destacar las plataformas digitales de aprendizaje (con una gran diferencia), seguido de sitios/recursos y referencias para el aprendizaje.

Ahora, veamos qué sector son los que más utilizan este tipo de herramientas.

```python
plt.figure(figsize=(16, 10))
B = sns.countplot(y='sector(s)',  #eje y
              data=products_df, #info gráfico
              order=products_df["sector(s)"].value_counts().index[:10], palette="husl") #info gráfico
B.set(xlabel="Suma", ylabel = "sector(s)") #etiquetas ejes
plt.title("Cantidad por sectores")
plt.show()
```

Podemos ver que el sector que más utiliza las herramientas propuestas es Prekinder junto con educación superior.

Esto nos entregó mucha información sobre los datos a analizar, y sobre todo, el enfoque de enseñanza que se tuvo durante la pandemia, donde las compañías rápidamente se posicionaron para la ayuda del estudiantado, brindando múltiples herramientas para proseguir con el estudio, sobre todo, fue una herramienta importante para cursos de pre-kinder.
