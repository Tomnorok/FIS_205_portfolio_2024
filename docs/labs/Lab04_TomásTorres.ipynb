EL conjunto de datos se denomina `company_sales_data.csv`, el cual contiene información tal como: número del mes, unidades, precio, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_04/data/company_sales_data.csv")
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

Para el gráfico lineal se tendrá:

```python
eje_x = df['month_number'] #Definimos eje x
eje_y = df['total_profit'] #Define eje y

plt.plot(eje_x, eje_y) #Ploteamos el gráfico
plt.title('Beneficio Total vs Mes') #Le ponemos nombre al gráfico
plt.xlabel('Mes') #y al eje x
plt.ylabel('Beneficio Total') #y al eje y
plt.show() #y lo mostramos
```

Ahora, para el gráfico de dispersión:

```python
data={'profit': eje_x, 'month': eje_y} #Creamos el eje x e y

dfa= pd.DataFrame(data = data); #Lo hacemos un DataFrame
dfa.plot.scatter(x = 'profit', y = 'month', s = 50, c='blue'); #Creamos el gráfico de dispersión
plt.title('Beneficio Total vs Mes') #Le ponemos nombre al gráfico
plt.xlabel('Mes') #y al eje x
plt.ylabel('Beneficio Total') #y al eje y
plt.show() #y lo mostramos
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
for col in df.columns: #Leemos cada columna
    if not col == 'month_number' and not col == 'total_profit' and not col =='total_units': #ponemos condiciones para no leer total units ni total profit
        plt.plot(df['month_number'], df[col], label=col) #ploteamos las ventas de la columna

plt.legend() #Se imprime la leyenda de cada uno de los productos
plt.title('Ventas según producto') #Le ponemos nombre al gráfico
plt.xlabel('Mes') #y al eje x
plt.ylabel('Ventas') #y al eje y
plt.show() #y lo mostramos
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
df[['facecream', 'facewash']].plot.bar() #Se grafican las barras que tangan de nombre 'facecream' y 'facewash' (nombre de las columnas del df)
plt.title('Ventas según producto por mes') #Le ponemos nombre al gráfico
plt.xlabel('Mes') #y al eje x
plt.ylabel('Ventas') #y al eje y
plt.show() #y lo mostramos
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
df1=df.drop(['month_number','total_units','total_profit'], axis=1) #Creamos un df sin las columnas innecesarias
sns.boxplot(data=df1) #Tipo de grafico de caja
plt.title('Ventas según producto') #Le ponemos nombre al gráfico
plt.xlabel('Producto') #y al eje x
plt.ylabel('Ventas') #y al eje y
plt.show() #y lo mostramos
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
J=df1[df1.columns[0:]].sum() #Calcula la suma de todos los productos durante el año
plt.pie(J, labels=df1.columns, autopct='%1.1f%%') #Muestra el porcentaje de venta de cada producto en un gráfico de pie
plt.title('Ventas totales del año pasado por producto') #Título del gráfico
plt.show() #y lo mostramos
```
